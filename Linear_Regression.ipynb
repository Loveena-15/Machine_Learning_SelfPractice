Linear Regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3]])
Y = np.array([2, 4, 5])

model = LinearRegression()
model.fit(X, Y)

print("Slope:", model.coef_)
print("Intercept:", model.intercept_)

# prediction
print(model.predict([[4]]))
```

Errors- MSA.MAE,RMSA

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

y_true = [2, 4, 6]
y_pred = [3, 5, 7]

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
```

Another Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 1. Load dataset
data = pd.read_csv("house_data.csv")

# 2. Separate input and output
X = data[['Size']]
Y = data['Price']

# 3. Create model
model = LinearRegression()

# 4. Train model
model.fit(X, Y)

# 5. Get slope and intercept
m = model.coef_[0]
b = model.intercept_

print("Slope (m):", m)
print("Intercept (b):", b)

# 6. Prediction
size_input = [[2000]]
predicted_price = model.predict(size_input)

print("Predicted price for 2000 sq ft:", predicted_price[0])

# 7. Predictions on dataset
y_pred = model.predict(X)

# 8. Evaluation
mse = mean_squared_error(Y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)

# 9. Plot graph
plt.scatter(X, Y)  # actual points
plt.plot(X, y_pred)  # best fit line
plt.xlabel("Size")
plt.ylabel("Price")
plt.title("Linear Regression - House Price")
plt.show()
```
