KNN--> K nearest neighbours

Classification

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
# Data (X = features, y = labels)
X = [[1, 2], [2, 3], [3, 3], [6, 7], [7, 8]]
y = [0, 0, 0, 1, 1]  # 0 = Class A, 1 = Class B

Scaler=StandardScaler()
X_Scaled=Scaler.fit_transform(X)
# Model
knn = KNeighborsClassifier(n_neighbors=3)

# Train
knn.fit(X_Scaled, y)

# New point
query = [[5, 5]]

# Get neighbors
distances, indices = knn.kneighbors(query)

print("Indices:", indices)
print("Distances:", distances)

# Actual points
neighbors = [X[i] for i in indices[0]]
labels = [y[i] for i in indices[0]]

print("Neighbor points:", neighbors)
print("Neighbor labels:", labels)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
# Data
X = [[1], [2], [3], [4], [5]]
y = [2, 4, 6, 8, 10]
Scaler=StandardScaler()
X_Scaled=Scaler.fit_transform(X)
# Model
knn = KNeighborsRegressor(n_neighbors=2)

# Train
knn.fit(X_Scaled, y)

# Query point
query = [[3.5]]

# Step 1: Prediction
prediction = knn.predict(query)
print("Prediction:", prediction)

# Step 2: Find neighbors
distances, indices = knn.kneighbors(query)

print("Indices:", indices)
print("Distances:", distances)

# Step 3: Actual neighbors
neighbors_X = [X[i] for i in indices[0]]
neighbors_y = [y[i] for i in indices[0]]

print("Neighbor points:", neighbors_X)
print("Neighbor values:", neighbors_y)
```

Weighted KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
# Data
X = [[1000], [1100], [1200], [2000]]
y = [20, 22, 24, 50]
Scaler=StandardScaler()
X_Scaled=Scaler.fit_transform(X)
# Weighted KNN
knn = KNeighborsRegressor(n_neighbors=3, weights='distance')

knn.fit(X_Scaled, y)

# Prediction
print(knn.predict([[1150]]))
```
