**Dimensionality Reduction**

PCA- Principal Component Analysis

EVD---> Eigen Value Decomposition

```python
from sklearn.decomposition import PCA
import numpy as np

# Data
X = np.array([
    [2, 4],
    [4, 2],
    [6, 8]
])

# PCA object
pca = PCA(n_components=1)

# Fit + Transform
X_reduced = pca.fit_transform(X)

print("Reduced Data:\n", X_reduced)

print("\nExplained Variance:", pca.explained_variance_)
print("\nPrincipal Component:", pca.components_)
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

data = load_iris()
X = data.data

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

print(X_reduced[:5])
```

SVD ---> Singular Value Decomposition

```python
from sklearn.decomposition import TruncatedSVD

X = np.random.rand(100, 10)

svd = TruncatedSVD(n_components=2)
X_reduced = svd.fit_transform(X)

print(X_reduced.shape)
print("Components (V):\n", svd.components_)
print("Explained variance:\n", svd.explained_variance_)
print("Explained variance ratio:\n", svd.explained_variance_ratio_)
```

t-SNE (t-distribution stochastic neighbourhood distribution)

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load data
data = load_digits()
X = data.data
y = data.target

# Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000)
X_embedded = tsne.fit_transform(X)

# Plot
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y)
plt.colorbar()
plt.show()
```
