Batch Gradient descent

```python
import numpy as np

# 🔹 1. Sample Dataset (y = 2x + 3)
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([5, 7, 9, 11, 13], dtype=float)

# 🔹 2. Initialize parameters
w = 0.0   # weight
b = 0.0   # bias

# 🔹 3. Hyperparameters
learning_rate = 0.01
epochs = 10

n = len(X)

# 🔹 4. Training using Gradient Descent
for epoch in range(epochs):

    y_pred = w * X + b   # model prediction

    # Loss (Mean Squared Error)
    loss = np.mean((y - y_pred) ** 2)

    # Gradients
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print progress
    print(f"Epoch {epoch+1}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")

# 🔹 Final Result
print("\nFinal Parameters:")
print(f"w = {w}, b = {b}")
```

Stochastic Gradient  Descent

```python
import numpy as np

# 🔹 1. Dataset (y = 2x + 3)
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([5, 7, 9, 11, 13], dtype=float)

# 🔹 2. Model parameters
w = 0.0
b = 0.0

# 🔹 3. Hyperparameters
learning_rate = 0.01
epochs = 10

# 🔹 4. Model function
def model(x):
    return w * x + b

# 🔹 5. Loss function (MSE for single point)
def compute_loss(y_true, y_pred):
    return (y_true - y_pred) ** 2

# 🔹 6. Gradient function
def compute_gradient(x, y_true, y_pred):
    error = y_true - y_pred
    dw = -2 * x * error
    db = -2 * error
    return dw, db

# 🔹 7. Training (SGD)
for epoch in range(epochs):

    total_loss = 0

    for i in range(len(X)):

        x_i = X[i]
        y_i = y[i]

        # Prediction
        y_pred = w * x_i + b

        # Loss
        loss = compute_loss(y_i, y_pred)
        total_loss += loss

        # Gradient
        dw, db = compute_gradient(x_i, y_i, y_pred)

        #  Immediate update (SGD key step)
        w = w - learning_rate * dw
        b = b - learning_rate * db

        # Debug print (optional)
        print(f"Epoch {epoch+1}, Sample {i+1}: w={w:.4f}, b={b:.4f}, loss={loss:.4f}")

    avg_loss = total_loss / len(X)
    print(f"\n Epoch {epoch+1} completed | Avg Loss = {avg_loss:.4f}\n")

# 🔹 Final Result
print("Final Parameters:")
print(f"w = {w:.4f}, b = {b:.4f}")
```
