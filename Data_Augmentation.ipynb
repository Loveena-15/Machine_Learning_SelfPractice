1.) Tabular Data

A.) Noise Injection

```python
import numpy as np

# original data
X = np.array([1000, 1200, 1500])

# generate noise
noise = np.random.normal(0, 20, X.shape)  # mean=0, std=20

# new data
X_new = X + noise

print("Original:", X)
print("Noisy:", X_new)
```

B.) SMOTE-(Synthetic Minority Over-Sampling)

```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt

# Step 2: Create an imbalanced dataset
X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_clusters_per_class=1,
    weights=[0.9, 0.1],  # 90% class 0, 10% class 1
    random_state=42
)

print("Before SMOTE:", Counter(y))

# Step 3: Apply SMOTE
smote = SMOTE(
    sampling_strategy='auto',  # balance classes
    k_neighbors=5,
    random_state=42
)

X_resampled, y_resampled = smote.fit_resample(X, y)

print("After SMOTE:", Counter(y_resampled))

# Step 4: Visualize before and after
plt.figure(figsize=(10,4))

# Before
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Before SMOTE")

# After
plt.subplot(1,2,2)
plt.scatter(X_resampled[:,0], X_resampled[:,1], c=y_resampled)
plt.title("After SMOTE")

plt.show()
```
